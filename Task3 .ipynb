```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('SampleSuperstore.csv')
data
```

```python
data.shape
```

```python
data.info()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

data['Ship_Mode_n'] = le.fit_transform(data['Ship Mode'])
```

```python
data['Segment_n'] = le.fit_transform(data['Segment'])
data['City_n'] = le.fit_transform(data['City'])
data['State_n'] = le.fit_transform(data['State'])
data['Region_n'] = le.fit_transform(data['Region'])
data['Category_n'] = le.fit_transform(data['Category'])
data['Sub-Cat_n'] = le.fit_transform(data['Sub-Category'])
```

```python
data
```

```python
data = data.drop(['Postal Code','Discount','Category','Segment','Ship Mode','Country','City','State','Sub-Category','Region'], axis='columns')
```

```python
data
```

```python
output = data['Profit']
data = data.drop(['Profit'],axis='columns')
data['Profit'] = output
data
```
